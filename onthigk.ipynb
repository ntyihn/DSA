#### __SELECTION SORT__ 

Duyệt qua danh sách, tìm phần tử nhỏ nhất trong danh sách chưa sắp xếp.

Hoán đổi phần tử nhỏ nhất đó với phần tử đầu tiên của danh sách chưa sắp xếp.

Lặp lại quá trình cho đến khi toàn bộ danh sách được sắp xếp.

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:  
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i] 

arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print(arr)  # Kết quả: [11, 12, 22, 25, 64]
```

#### __BUBBLE SORT__ (nếu mảng đã sắp xếp)

Lặp qua danh sách và so sánh từng cặp phần tử liên tiếp, nếu phần tử trước lớn 

hơn phần tử sau thì hoán đổi.

Sau mỗi vòng lặp, phần tử lớn nhất sẽ "nổi" lên cuối danh sách.

Lặp lại quá trình cho đến khi danh sách được sắp xếp.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False  # Biến kiểm tra xem có hoán đổi không
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:  # Hoán đổi nếu sai thứ tự
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:  # Nếu không có hoán đổi, mảng đã được sắp xếp
            break

arr = [64, 25, 12, 22, 11]
bubble_sort(arr)
print(arr)  # Kết quả: [11, 12, 22, 25, 64]
```

#### __INSERTION SORT__ (nếu mảng gần đúng)

Chia danh sách thành phần đã sắp xếp và phần chưa sắp xếp.

Lấy từng phần tử trong danh sách chưa sắp xếp và chèn vào đúng vị trí 

trong danh sách đã sắp xếp.

Tiếp tục cho đến khi toàn bộ danh sách được sắp xếp.

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:  # Dịch chuyển phần tử lớn hơn sang phải
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # Chèn phần tử vào đúng vị trí

arr = [64, 25, 12, 22, 11]
insertion_sort(arr)
print(arr)  # Kết quả: [11, 12, 22, 25, 64]
```

**Khi lớp con muốn kế thừa các thuộc tính của lớp cha phải làm thế nào?**

Khi một lớp con (child class) muốn kế thừa các thuộc tính của lớp cha

 (parent class), ta sử dụng super() để gọi hàm khởi tạo (__ init __) của lớp cha.

CÁCH KẾ THỪA: Dùng super().__ init __(tham_số_cần_truyền) để kế thừa constructor của lớp cha.

Lớp con có thể thêm thuộc tính hoặc phương thức mới mà lớp cha không có.



```python

class Person:
    def __init__(self, name, age):
        self.name = name  
        self.age = age  

    def introduce(self):
        return f"My name is {self.name} and I am {self.age} years old."

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)  
        self.student_id = student_id 

    def introduce(self):  # Ghi đè phương thức introduce của lớp cha
        return f"My name is {self.name}, I am {self.age} years old, and my student ID is {self.student_id}."

# Test
student = Student("Alice", 20, "S12345")
print(student.introduce())
 
# Output: "My name is Alice, I am 20 years old, and my student ID is S12345."
```

```python
class Media:
    def __init__(self, id, title, year, category, available=True):
        self.id = id
        self.title = title
        self.year = year
        self.category = category
        self.available = available
    
    def __str__(self):
        return f"ID: {self.id}, Title: {self.title}, Year: {self.year}, \
            Category: {self.category}, Available: {self.available}"
```

```python
class QLStack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        if len(self.stack) == 0:
            return True
        return False
    
    def push(self, number):
        self.stack.insert(0, number)
        
    def pop(self):
        return self.stack.pop(0)
```

```python
class QLQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0
    def push(self, number):
        self.stack.append(number)
    def pop(self):
        return self.stack.pop()
```
