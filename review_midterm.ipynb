```python
class Employee:
    def __init__(self, id, ten, tuoi, email, luongcoban):
        self.id = id
        self.ten = ten
        self.tuoi = tuoi
        self.email = email
        self.luongcoban = luongcoban
        
    def __str__(self):
        return f"ID: {self.id}, Tên: {self.ten}, Tuổi: {self.tuoi}, \
            Email: {self.email}, Lương cơ bản: {self.luongcoban}" 
        
```

```python
class Developer(Employee):
    def __init__(self, id, ten, tuoi, email, luongcoban, ngonngu, namkn, capdo):
        super().__init__(id, ten, tuoi, email, luongcoban)
        self.ngonngu = ngonngu
        self.namkn = namkn
        self.capdo = capdo
    
    def __str__(self):
        return f"{super().__str__()}, Ngôn ngữ: {self.ngonngu}, \
            Năm kinh nghiệm: {self.namkn}, Cấp độ: {self.capdo}"
```

```python
class ProjectManager(Employee):
    def __init__(self, id, ten, tuoi, email, luongcoban, soduan, chuyenmon, chungchi):
        super().__init__(self, id, ten, tuoi, email, luongcoban)
        self.soduan = soduan
        self.chuyenmon = chuyenmon
        self.chungchi = chungchi
    
    def __str__(self):
        return f"{super().__str__()}, Số dự án: {self.soduan}, \
            Chuyên môn: {self.chuyenmon}, Chứng chỉ: {self.chungchi}"
        
```

```python
class Tester(Employee):
    def __init__(self, id, ten, tuoi, email, luongcoban, loaikiemthu, congcusd):
        super().__init__(id, ten, tuoi, email, luongcoban)
        self.loaikiemthu = loaikiemthu
        self.congcusd = congcusd
    
    def __str__(self):
        return f"{super().__str__()}, Loại kiểm thử: {self.loaikiemthu}, \
            Công cụ sử dụng: {self.congcusd}"
        
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, item):
        self.stack.insert(0, item)
    def pop(self):
        return self.stack.pop(0)
    def is_empty(self):
        return len(self.stack) == 0
    def show(self):
        for i in self.stack:
            print(i)
```

```python
class EmployeeManagement:
    def __init__(self):
        self.ds = Stack()

# Thêm các loại nhân viên khác nhau vào hệ thống
    def add_employee(self, employee):
        self.ds.push(employee)

# Xóa nhân viên 
    def remove_employee(self):
        if self.ds.is_empty():
            return None
        return self.ds.pop()

# Hiển thị danh sách nhân viên 
    def show_employee(self):
        if self.ds.is_empty():
            return None
        for employee in reversed(self.ds.items):  
            print(employee)

# Tìm kiếm nhân viên theo ID 
    def search_employee(self, id): 
        for employee in self.ds.items:
            if employee.id == id:
                return True
        return False 
    
# Cập nhật thông tin nhân viên
    def update_employee(self, id, ten, tuoi, email, luongcoban):
        kq = self.search_employee(id)
        if kq == True:
            if id:
                kq.id = id
            if ten:
                kq.ten = ten
            if tuoi:
                kq.tuoi = tuoi
            if email:
                kq.email = email
            if luongcoban:
                kq.luongcoban = luongcoban
            return True
        return False

# Tìm nhân viên có kinh nghiệm cao nhất 
    def find_max_exp(self):
        max_exp = 0
        for employee in self.ds.items:
            if employee.namkn > max_exp:
                max_exp = employee.namkn
                max_exp_employee = employee
        return max_exp_employee

# Tìm quản lý dự án có nhiều dự án thành công nhất
    def find_max_project(self):
        max_project = 0
        for employee in self.ds.items:
            if employee.soduan > max_project:
                max_project = employee.soduan
                max_project_employee = employee
        return max_project_employee

# Sắp xếp nhân viên theo mức lương sử dụng Bubble Sort 
    def sort_employee(self):
        n = len(self.ds.items)
        for i in range(n):
            for j in range(n-i-1):
                if self.ds.items[j].luongcoban < self.ds.items[j+1].luongcoban:
                    self.ds.items[j], self.ds.items[j+1] = self.ds.items[j+1], self.ds.items[j]
        return self.ds.items
    
# Triển khai chức năng hoàn tác (undo) các thao tác quản lý nhân viên bằng Stack 
    def undo(self):
        if self.ds.is_empty():
            return None
        return self.ds.pop()

# Lưu dữ liệu nhân viên vào file 
    def save(self, filename = "Employee.txt"):
        with open(filename, "w", encoding= 'utf-8') as f:
            for i in self.ds.stack:
                f.write(str(i.show()))

# Đọc dữ liệu dự án và nhiệm vụ từ file


    
```

```python
def find_most_experienced_developer(self):
        developers = [emp for emp in self.employees if isinstance(emp, Developer)]
        if not developers:
            print("Không có lập trình viên trong hệ thống")
            return None
        
        most_experienced = max(developers, key=lambda dev: dev.years_experience)
        return most_experienced
    
def find_most_successful_manager(self):
        managers = [emp for emp in self.employees if isinstance(emp, ProjectManager)]
        if not managers:
            print("Không có quản lý dự án trong hệ thống")
            return None
        
        most_successful = max(managers, key=lambda mgr: mgr.completed_projects)
        return most_successful
```
